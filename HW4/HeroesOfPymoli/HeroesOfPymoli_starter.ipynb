```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
uniquePlayers = pd.DataFrame({"Total Players":[purchase_data.SN.nunique()]})
uniquePlayers
```

```python
uniqueItems = [purchase_data["Item ID"].nunique()]
avgPrice = [purchase_data["Price"].mean()]
numPurchase = [purchase_data["Purchase ID"].nunique()]
revenue = [purchase_data["Price"].sum()]
summary = pd.DataFrame({"# of Unique Items":uniqueItems,
                        "Average Price":avgPrice, 
                       "# of Purchases":numPurchase,
                       "Total Revenue":revenue})
summary["Average Price"] = summary["Average Price"].map('${:,.2f}'.format)
summary["Total Revenue"] = summary["Total Revenue"].map('${:,.2f}'.format)
summary
```

```python
genderDemo = purchase_data.drop_duplicates(subset = 'SN')
genderDemo = genderDemo.groupby("Gender").count()
genderDemo = genderDemo.drop(columns=['SN','Age', "Item ID", "Item Name", "Price"])
genderDemo.columns = ['Total Count']
genderDemo["Percentage of Players"] = (genderDemo['Total Count']/genderDemo['Total Count'].sum()*100).map('{:,.2f}%'.format)
genderDemo
```

```python
genderAnalysis = purchase_data.groupby(["Gender"])
genderAnalysisDF = pd.DataFrame({'Purchase Count':genderAnalysis["Purchase ID"].count(), 
                                 'Average Purchase Price':genderAnalysis["Price"].mean(),
                                 'Total Purchase Value':genderAnalysis["Price"].sum()})
genderAnalysisDF['Avg Total Purchase per Person'] = (genderAnalysisDF['Total Purchase Value']/genderDemo['Total Count']).map('${:,.2f}'.format)
genderAnalysisDF['Average Purchase Price'] = genderAnalysisDF['Average Purchase Price'].map('${:,.2f}'.format)
genderAnalysisDF['Total Purchase Value'] = genderAnalysisDF['Total Purchase Value'].map('${:,.2f}'.format)
genderAnalysisDF
```

```python
noDup = purchase_data.drop_duplicates(subset = 'SN')
bins = [0,9,14,19,24,29,34,39,noDup['Age'].max()]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
noDup['Age Group'] = pd.cut(noDup['Age'], bins, labels=labels)

ageGroup = noDup.groupby('Age Group')
ageGroupDF = pd.DataFrame({'Count':ageGroup['SN'].count()})
ageGroupDF["% of Players"]= (ageGroupDF['Count']/ageGroupDF['Count'].sum()*100).map('{:,.2f}%'.format)
ageGroupDF
```

```python
bins = [0,9,14,19,24,29,34,39,purchase_data['Age'].max()]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], bins, labels=labels)
purchaseGroup = purchase_data.groupby('Age Group')
ageAnalysis = pd.DataFrame({
    'Purchase Count':purchaseGroup['SN'].count(),
    'Average Purchase Price':purchaseGroup['Price'].mean(),
    'Total Purchase Value':purchaseGroup['Price'].sum()})
ageAnalysis['Avg Purchase/Person'] = (ageAnalysis['Total Purchase Value']/ageAnalysis['Purchase Count']).map('${:,.2f}'.format)
ageAnalysis['Total Purchase Value'] = ageAnalysis['Total Purchase Value'].map('${:,.2f}'.format)
ageAnalysis['Average Purchase Price'] = ageAnalysis['Average Purchase Price'].map('${:,.2f}'.format)
ageAnalysis
```

```python
SNgroup = purchase_data.groupby('SN')
SNanalysis = pd.DataFrame({
    'Purchase Count':SNgroup['Purchase ID'].count(),
    'Average Purchase Price':SNgroup['Price'].mean(),
    'Total Purchase Value':SNgroup['Price'].sum()})
SNanalysis = SNanalysis.sort_values(by=['Total Purchase Value'], ascending = False)
SNanalysis['Average Purchase Price']= SNanalysis['Average Purchase Price'].map('${:,.2f}'.format)
SNanalysis['Total Purchase Value']= SNanalysis['Total Purchase Value'].map('${:,.2f}'.format)
SNanalysis.head()
```

```python
itemGroup = purchase_data.groupby(['Item ID','Item Name'])
itemAnalysis = pd.DataFrame({'Purchase Count':itemGroup['Purchase ID'].count()})
ID_Price = purchase_data[['Item ID','Item Name','Price']].set_index('Item ID','Item Name')
itemAnalysis['Total Purchase Value']=itemGroup['Price'].sum()
itemAnalysis = itemAnalysis.merge(ID_Price,on = ["Item ID",'Item Name'], how = 'left')
itemAnalysis = itemAnalysis.sort_values(by=['Purchase Count'], ascending = False)
itemAnalysis = itemAnalysis.drop_duplicates(subset = 'Item Name')
itemAnalysis['Price']= itemAnalysis['Price'].map('${:,.2f}'.format)
itemAnalysis = itemAnalysis[['Item Name','Purchase Count','Price','Total Purchase Value']]
#itemAnalysis['Total Purchase Value']= itemAnalysis['Total Purchase Value'].map('${:,.2f}'.format)
itemAnalysis.head()
```
