```python
from sqlalchemy import create_engine
import databaseconfig as cfg
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
engine = create_engine('postgresql://postgres:'+cfg.mysql['password']+'@localhost:5432/Homework')
connection = engine.connect()
```

```python
df = pd.read_sql("SELECT title, ROUND(AVG(s.salary), 2)FROM titles t JOIN salaries s ON t.emp_no=s.emp_no GROUP BY title;", connection)
y_pos = np.arange(len(df['round']))
y_pos
```

```python
plt.bar(y_pos, df['round'], align='center', alpha=0.5)
plt.xticks(y_pos, df['title'], rotation=90)
plt.ylabel('Salary ($)')
plt.title('Average Salary of Positions')
plt.savefig('Average Salary.png')
```
